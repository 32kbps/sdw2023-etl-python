```python
df.columns = ["Nome", "Conta", "Cartao"]
df

```

```python
users = df.to_dict(orient="records")

def gerar_mensagem(user):
    nome = user["Nome"]
    mensagem = f"{nome}, investir hoje é o primeiro passo para um futuro financeiro melhor."
    return mensagem[:100]

for u in users:
    u["Mensagem"] = gerar_mensagem(u)

users

```

```python
out_df = pd.DataFrame(users)
out_df.to_csv("SDW2023_out.csv", index=False, encoding="utf-8")

out_df

```
